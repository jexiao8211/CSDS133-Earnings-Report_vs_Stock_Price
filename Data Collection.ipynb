# Get Quarterly Finances 
### getQFinances function
The getQFinances function (seen below) will return the last 4 quarterly finance reports for any given ticker (for example try seraching for the AAPL finacnes).

Make sure the ticker is all caps and a string with no spaces

### What data is availible
I can get the last 4 quarters of information which includes the following highlights
<ul>
    <li> Income Before Tax
    <li> Gross Income
    <li> Total Revenue
    <li> Net income (use net income applicable to common shares)
    <li> Cost of Revenue
    <li> R&D cost
    <li> EBIT
</ul>
<strong>Note</strong> that we may want to use earnings per share to account for stock buybacks or issuing. . .we need to calculate ourselves :( text me if you want me to code that

Data is retunred as a DataFrame so make sure to use .loc([row, column]) to select data
    

```python
def getQFinances(ticker):
    if type(ticker) != str:
        print("pls make ticker in a string format")
        return
    return(yf.Ticker(ticker).quarterly_financials)
    
```

### Getting Finance per earnings report

getFin(ticker, indicator)   --> both ticker and indicator are strings
ticker is what stock you want
indicator is what thing you want to measure change (for instance "Total Revenue")

It will return the last 4 quarters of that peice of data in an array, where index 0 is last quarter and index 3 is 4 quarters ago


```python
def getFin(ticker, indic):
    qFinance =   getQFinances(ticker)
    data = qFinance.loc[indic, :]
    return([data[0], data[1], data[2], data[3]])
    

def getGenFin(ticker):
    qFinance =   getQFinances(ticker)
    data = qFinance.loc["Total Revenue", :]
    revenue = [data[0], data[1], data[2], data[3]]    
    data = qFinance.loc["Cost Of Revenue", :]
    cogs = [data[0], data[1], data[2], data[3]] 
    data = qFinance.loc["Gross Profit", :]
    gp = [data[0], data[1], data[2], data[3]]
    return(revenue, cogs, gp)

getGenFin("INTC")
```

### These functions return last or next trading date
Works by making something into a date time object, checking for public holidays, then finding the next weekday

```python

def getLastTDay(day):
    if(type(day) == str):
        day = dt.datetime(int(day[6]+day[7]+day[8]+day[9]),int(day[0]+day[1]), int(day[3]+day[4]))
    weekday = int(day.strftime("%w")) #returns integer of what weekday it is      
    if(weekday == 0): #if sunday
        day = day + dt.timedelta(days = -2)
    if(weekday == 6): #if saturday
        day = day + dt.timedelta(days = -1)
    #@#@#@#@#@#@#@#@#@#@#@#@#@##@ Exception time @#@#@#@#@#@#@#@#@#@#@#@#@##@
    if(day.strftime("%m-%d-%Y") == "01-01-2021"): #New years
        day = day + dt.timedelta(days = -1)
       
    if(day.strftime("%m-%d-%Y") == "01-17-2022" or day.strftime("%m-%d-%Y") == "01-18-2021"): #MLK day exception
        day = day + dt.timedelta(days = -1)
   
    if(day.strftime("%m-%d-%Y") == "04-02-2021" or day.strftime("%m-%d-%Y") == "04-15-2022"): #Good friday
        day = day + dt.timedelta(days = -1)
   
    if(day.strftime("%m-%d-%Y") == "05-31-2021"): #Memorial Day
        day = day + dt.timedelta(days = -1)
       
    if(day.strftime("%m-%d-%Y") == "07-05-2021"): #Indpendence Day
        day = day + dt.timedelta(days = -1)        
       
    if(day.strftime("%m-%d-%Y") == "09-06-2021"): #Labor day
        day = day + dt.timedelta(days = -1)              
       
    if(day.strftime("%m-%d-%Y") == "11-25-2021"): #Thanksgiving
        day = day + dt.timedelta(days = -1)              
   
    if(day.strftime("%m-%d-%Y") == "12-24-2021"): #Christmas
        day = day + dt.timedelta(days = -1)      
   
    #@#@#@#@#@#@#@#@#@#@#@#@#@##@ Exceptions done @#@#@#@#@#@#@#@#@#@#@#@#@##@
   
    weekday = int(day.strftime("%w")) #returns integer of what weekday it is  
   
    if(weekday >0 and weekday <6): #checks if it mon-fri
        return day.strftime("%Y-%m-%d")
    if(weekday == 0): #if sunday
        day = day + dt.timedelta(days = -2)
        return day.strftime("%Y-%m-%d")
    if(weekday == 6): #if saturday
        day = day + dt.timedelta(days = -1)
        return day.strftime("%Y-%m-%d")
   

   
def getNextTDay(day):
    if(type(day) == str):
        day = dt.datetime(int(day[6]+day[7]+day[8]+day[9]),int(day[0]+day[1]), int(day[3]+day[4]))
    day = day + dt.timedelta(days = 1)
    weekday = int(day.strftime("%w")) #returns integer of what weekday it is 
    weekday = int(day.strftime("%w")) #returns integer of what weekday it is    
    if(weekday == 0):
        day = day + dt.timedelta(days = 1)
    if(weekday == 6):
        day = day + dt.timedelta(days = 2)
    #@#@#@#@#@#@#@#@#@#@#@#@#@##@ Exception time @#@#@#@#@#@#@#@#@#@#@#@#@##@
    if(day.strftime("%m-%d-%Y") == "01-01-2021"): #New years
        day = day + dt.timedelta(days = 1)
       
    if(day.strftime("%m-%d-%Y") == "01-17-2022" or day.strftime("%m-%d-%Y") == "01-18-2021"): #MLK day exception
        day = day + dt.timedelta(days = 1)
   
    if(day.strftime("%m-%d-%Y") == "04-02-2021" or day.strftime("%m-%d-%Y") == "04-15-2022"): #Good friday
        day = day + dt.timedelta(days = 1)
   
    if(day.strftime("%m-%d-%Y") == "05-31-2021"): #Memorial Day
        day = day + dt.timedelta(days = 1)
       
    if(day.strftime("%m-%d-%Y") == "07-05-2021"): #Indpendence Day
        day = day + dt.timedelta(days = 1)        
       
    if(day.strftime("%m-%d-%Y") == "09-06-2021"): #Labor day
        day = day + dt.timedelta(days = 1)              
       
    if(day.strftime("%m-%d-%Y") == "11-25-2021"): #Thanksgiving
        day = day + dt.timedelta(days = 1)              
   
    if(day.strftime("%m-%d-%Y") == "12-24-2021"): #Christmas
        day = day + dt.timedelta(days = 1)      
    #@#@#@#@#@#@#@#@#@#@#@#@#@##@ Exceptions done @#@#@#@#@#@#@#@#@#@#@#@#@##@    
   
    weekday = int(day.strftime("%w")) #returns integer of what weekday it is    
    if(weekday >0 and weekday <6):
        return day.strftime("%Y-%m-%d")
    if(weekday == 0):
        day = day + dt.timedelta(days = 1)
        return day.strftime("%Y-%m-%d")
    if(weekday == 6):
        day = day + dt.timedelta(days = 2)
        return day.strftime("%Y-%m-%d")
```

### getPrice returns the prices before and after earnings
it does this through returning two arrays, the first is before earings price and second is after earnings price. It uses close price and returns from last 4 quarters 

```python

def getPrice(ticker):
    
    stock = yf.Ticker(ticker) 
    data = stock.history(period = "2y") #get historical stock prices
    
    dates = [eDates.loc[ticker, "Recent"],eDates.loc[ticker, "Second Recent"],eDates.loc[ticker, "Third Recent"],eDates.loc[ticker, "Fourth Recent"]]
    beforeDates = [getLastTDay(dates[0]),getLastTDay(dates[1]),getLastTDay(dates[2]),getLastTDay(dates[3])]
    afterDates = [getNextTDay(dates[0]),getNextTDay(dates[1]),getNextTDay(dates[2]),getNextTDay(dates[3])]

    beforeEarnings = [data.loc[beforeDates[0],"Close"], data.loc[beforeDates[1],"Close"], data.loc[beforeDates[2],"Close"],data.loc[beforeDates[3],"Close"]]    
    afterEarnings = [data.loc[afterDates[0],"Close"], data.loc[afterDates[1],"Close"], data.loc[afterDates[2],"Close"],data.loc[afterDates[3],"Close"]]    

    return(beforeEarnings, afterEarnings)



test, test2 = getPrice("AAPL")

print(test)
print(test2)

```
