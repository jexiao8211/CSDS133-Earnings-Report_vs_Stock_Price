# IF R^2 IS NEGATIVE, MEANS THAT ITS A WORSE FIT THAN A HORIZONTAL LINE

## Import stock price info
- create a difference in stock price table (just for most recent before for now)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
prices = pd.read_csv(r'stockPrice.csv')
pricedif1 = {'price change(%)': ((prices.iloc[:, 2] - prices.iloc[:, 1])/prices.iloc[:, 1])*100}
pricedif1 = pd.DataFrame.from_dict(pricedif1)
pricedif1
```

# Putting data into one dataframe
- results are very different for each run through

```python
#Revenue
rev = pd.read_csv(r'revenue.csv')
revdiff = {'rev change %': ((rev.iloc[:, 1] - rev.iloc[:, 2])/rev.iloc[:, 2])*100}
revdata = pd.DataFrame.from_dict(revdiff).join(pricedif1)
data = revdata
#Net Income
income = pd.read_csv(r'netIncome.csv')
incomediff = {'income change %': ((income.iloc[:, 1] - income.iloc[:, 2])/income.iloc[:, 2])*100}
data = data.join(pd.DataFrame.from_dict(incomediff))
#Gross Profit
grossProfit = pd.read_csv('grossProfit.csv')
gpdiff = {'gp change %': ((grossProfit.iloc[:, 1] - grossProfit.iloc[:, 2])/grossProfit.iloc[:, 2])*100}
data = data.join(pd.DataFrame.from_dict(gpdiff))
#Suprise
suprise = pd.read_csv('earnings suprise.csv')
data = data.join(suprise.iloc[:,1])









data
```

# Correlation Tests
find pearsons and spearman correlation coeff and P-val

```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr

# *&*&*&*&*&*&*&*&*&*& Revenue *&*&*&*&*&*&*&*&*&*&

corr, pval = pearsonr(data["gp change %"].tolist(), data["price change(%)"].tolist())
print("Revenue v. Price:  the pearson pval is: " + str(pval) + "      the pearsons corr is: "+ str(corr))
corr, pval = spearmanr(data["rev change %"].tolist(), data["price change(%)"].tolist())
print("Revenue v. Price:  the spearman pval is: " + str(pval) + "      the spearman corr is: "+ str(corr))

# *&*&*&*&*&*&*&*&*&*& Gross Profit *&*&*&*&*&*&*&*&*&*&

print("")
corr, pval = pearsonr(data["gp change %"].tolist(), data["price change(%)"].tolist())
print("Gross Profit v. Price:  the pearson pval is: " + str(pval) + "      the pearsons corr is: "+ str(corr))
corr, pval = spearmanr(data["gp change %"].tolist(), data["price change(%)"].tolist())
print("Gross Profit v. Price:  the spearman pval is: " + str(pval) + "      the spearman corr is: "+ str(corr))

# *&*&*&*&*&*&*&*&*&*& Net Income *&*&*&*&*&*&*&*&*&*&

print("")
corr, pval = pearsonr(data["income change %"].tolist(), data["price change(%)"].tolist())
print("Net Income v. Price:  the pearson pval is: " + str(pval) + "      the pearsons corr is: "+ str(corr))
corr, pval = spearmanr(data["income change %"].tolist(), data["price change(%)"].tolist())
print("Net Income v. Price:  the spearman pval is: " + str(pval) + "      the spearman corr is: "+ str(corr))

# *&*&*&*&*&*&*&*&*&*& Suprise *&*&*&*&*&*&*&*&*&*&

print("")
corr, pval = pearsonr(data["Most Recent Suprise"].tolist(), data["price change(%)"].tolist())
print("Suprise v. Price:  the pearson pval is: " + str(pval) + "      the pearsons corr is: "+ str(corr))
corr, pval = spearmanr(data["Most Recent Suprise"].tolist(), data["price change(%)"].tolist())
print("Suprise v. Price:  the spearman pval is: " + str(pval) + "      the spearman corr is: "+ str(corr))
```

# KNN model 


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
#split dataset into train and test data

X = data.drop(columns=['price change(%)'])
y = []

for price in data["price change(%)"].tolist():
    if(price>5):
        y.append(4)
    elif(price>0):
        y.append(3)
    elif(price>-5):
        y.append(2)
    else:
        y.append(1)
        
knn = KNeighborsClassifier(n_neighbors=1)


knnScores = []

for x in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, stratify=y)
    knn.fit(X_train,y_train)
    knnScores.append(knn.score(X_test, y_test))
knnScores[0:10]

```

```python
mean = sum(knnScores) / len(knnScores)
variance = sum([((x - mean) ** 2) for x in knnScores]) / len(knnScores)
res = variance ** 0.5
print("The mean is " + str(mean))
print("The standard devaiation is "+str(res))

```

# Polynomial Regression

```python
from sklearn.metrics import r2_score

polyModel = np.poly1d(np.polyfit(data["Most Recent Suprise"].tolist(), data["price change(%)"].tolist(), 5))
polyLine = np.linspace(-20, 75, 400)

plt.scatter(data["Most Recent Suprise"].tolist(), data["price change(%)"].tolist())
plt.plot(polyLine, polyModel(polyLine))
plt.show()

print("the r^2 value is: "+str(r2_score(data["price change(%)"].tolist(), polyModel(data["Most Recent Suprise"].tolist()))))
```
